# 2.1.1 Пример 1
* постройте диаграмму рассеяним sleep vs totwrk с ‘подогнанной’ прямой
* найти параметры оптимальной прямой sleep на totwrk

$_Подключим необходимые библиотеки в файле OLS_$

```python
import numpy as np
import pandas as pd

from statsmodels.formula.api import ols

import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('sleep75 (1).csv')
# размер датафрейма
df.shape
```

$_Строим диаграмму и находим параметры _$

```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='totwrk', y='sleep', ci=None, scatter_kws={'s':5}, line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# спецификация модели через формулу + найти параметры оптимальной прямой sleep на totwrk
mod = ols(formula='sleep~1+totwrk', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

```python
# найти параметры оптимальной прямой sleep на totwrk без константы
# тоже самое но 1 --> 0
mod = ols(formula='sleep~0+totwrk', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

# 2.1.2 Пример 2
* 1. постройте диаграмму рассеяним sleep vs age с ‘подогнанной’ прямой
* 2. найти параметры оптимальной прямой sleep на age
* 3. найти параметры оптимальной прямой sleep на age без константы

```python
# 1
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='age', y='sleep', ci=None, scatter_kws={'s':5}, line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# 2 спецификация модели через формулу + найти параметры оптимальной прямой sleep на age
mod = ols(formula='sleep~1+age', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

```python
# 3. найти параметры оптимальной прямой sleep на age без константы
# тоже самое но 1 --> 0
mod = ols(formula='sleep~0+age', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

# 2.1.3 Пример 3
* 1 постройте диаграмму рассеяним sleep vs totwrk с ‘подогнанной’ параболой
* 2 найти параметры оптимальной параболы sleep на totwrk, totwrk2

```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='totwrk', y='sleep', order=2,  # порядок полинома (2 = парабола), 
            ci=None, 
            scatter_kws={'s':5}, 
            line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# 1 спецификация модели через формулу + найти параметры оптимальной прямой sleep на ttwrk, ttwrk2

mod = ols(formula='sleep~1+totwrk+I(totwrk ** 2)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

$_ для squrt используй I(parametr**2)!!!_$

# 2.1.4 Пример 4
* 1 постройте диаграмму рассеяним sleep vs age с ‘подогнанной’ параболой
$_ появилась в файле артамонова_$
* 2 найти параметры оптимальной параболы sleep на age, age2


```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='age', y='sleep', order=2,  # порядок полинома (2 = парабола), 
            ci=None, 
            scatter_kws={'s':5}, 
            line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# спецификация модели через формулу
mod = ols(formula='sleep~1+age+I(age**2)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```
