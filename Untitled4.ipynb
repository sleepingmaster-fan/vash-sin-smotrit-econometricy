```python
import numpy as np
import pandas as pd

from statsmodels.formula.api import ols

import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('sleep75 (1).csv')
# размер датафрейма
df.shape
```

```python
#1.1 
# спецификация модели через формулу
mod = ols(formula='sleep~1+totwrk+male', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

Интерпретация коэффициентов
Константа (3573.20): Базовый уровень сна для женщины, которая не работает (0 минут). Прогноз — 3573 минуты сна в неделю.

totwrk (-0.17): Дополнительная минута работы снижает время сна на 0,17 мин. (≈10 сек.) при неизменном поле.

male (88.84): Мужчины спят в среднем на 88,8 минут в неделю больше, чем женщины при одинаковом количестве рабочих часов.



```python
#1.2
# спецификация модели через формулу
mod = ols(formula='sleep~1+totwrk+male+smsa+age+south+yngkid+marr+union', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params.round(2)
```

нтерпретация (Модель 2)
totwrk (-0.17): При прочих равных, дополнительная минута работы сокращает сон на 0,17 мин. (≈10 сек.).

male (87.11): При прочих равных, мужчины спят на 87 минут в неделю больше женщин.

Коэффициенты практически не изменились относительно Модели 1, даже после добавления контроля.



```python
#1.3 
# спецификация модели через формулу
mod = ols(formula='sleep~1+np.log(hrwage)+smsa+totwrk+male+marr+age+yngkid+south', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params.round(2)
```

log(hrwage) (-1.39): При росте зарплаты на 1% сон уменьшается на 0.0139 мин. (эффект практически нулевой).

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('Labour.csv')
# размер датафрейма
df.shape
```

```python
#3.1 
# спецификация модели через формулу
mod = ols(formula='np.log(output)~1+np.log(capital)+np.log(labour)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

log(capital) (0.208): При увеличении капитала на 1% (и неизменном труде) выпуск продукции увеличивается на 0.208%. Это эластичность выпуска по капиталу.

log(labour) (0.715): При увеличении труда на 1% (и неизменном капитале) выпуск продукции увеличивается на 0.715%. Это эластичность выпуска по труду.

*Примечание: сумма коэффициентов (0.208 + 0.715 = 0.923) меньше 1, что указывает на убывающую отдачу от масштаба.

```python
#3.2 
# спецификация модели через формулу
mod = ols(formula='np.log(output)~1+np.log(capital)+np.log(labour)+np.log(wage)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

log(capital) (0.149): При увеличении капитала на 1% (и фиксированных труде и зарплате) выпуск увеличивается на 0.149%.

log(labour) (0.720): При увеличении труда на 1% (и фиксированных капитале и зарплате) выпуск увеличивается на 0.720%.

Примечание: Сумма эластичностей выпуска по капиталу и труду равна 0.869, что означает убывающую отдачу от масштаба.

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('Electricity.csv')
# размер датафрейма
df.shape
```

```python
#4.1 
# спецификация модели через формулу
mod = ols(formula='np.log(cost)~1+np.log(q)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

log(q) (0.836): При увеличении объема производства (q) на 1% (при прочих равных) издержки (cost) увеличиваются в среднем на 0.836%.

Экономическая интерпретация: Коэффициент 0.836 < 1 свидетельствует о наличии положительной экономии от масштаба (с ростом выпуска издержки растут медленнее, чем выпуск).

```python
#4.2
# спецификация модели через формулу
mod = ols(formula='np.log(cost)~1+np.log(q)+np.log(pl)+np.log(pf)+np.log(pk)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

og(q) (0.838): При увеличении объема производства на 1% (и фиксированных ценах на труд, топливо и капитал) издержки увеличиваются в среднем на 0.838%.

Коэффициент по-прежнему меньше 1, что указывает на положительную экономию от масштаба даже после контроля цен факторов производства.
