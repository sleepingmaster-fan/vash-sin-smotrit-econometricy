# 2.2.1 Пример 1
* постройте диаграмму рассеяним output vs capital с ‘подогнанной’ прямой
* найти параметры оптимальной прямой output на capital


```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='capital', y='output', ci=None, scatter_kws={'s':5}, line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# спецификация модели через формулу + найти параметры оптимальной прямой output на capital
mod = ols(formula='output~1+capital', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

* найти параметры оптимальной прямой output на capital без константы

```python
# спецификация модели через формулу 0!! + найти параметры оптимальной прямой output на capital
mod = ols(formula='output~0+capital', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

* постройте диаграмму рассеяним log(output) vs log(capital) с ‘подогнанной’ прямой
  
# $_ log(parametr)=np.log(df['parametr'])_$

```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x=np.log(df['capital']), y=np.log(df['output']), ci=None, scatter_kws={'s':5},  line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

* найти параметры оптимальной прямой log(output) на log(capital)

```python
# спецификация модели через формулу + найти параметры оптимальной прямой output на capital
mod = ols(formula='np.log(output)~1+np.log(capital)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

* найти параметры оптимальной прямой log(output) на log(capital) без константы


```python
# спецификация модели через формулу с нулем + найти параметры оптимальной прямой output на capital
mod = ols(formula='np.log(output)~0+np.log(capital)', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
res.params
```

# 2.2.3 Пример 3
* постройте диаграмму рассеяним log(output) vs log(capital) с ‘подогнанной’ параболой
* найти параметры оптимальной параболы log(output) vs log(capital), log2(capital)

```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x=np.log(df['capital']), y=np.log(df['output']), order=2,  # порядок полинома (2 = парабола), 
            ci=None, 
            scatter_kws={'s':5}, 
            line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# спецификация модели через формулу
mod = ols(formula='np.log(output)~1+np.log(capital)+(I(np.log(capital)**2))', data=df)
# подгонка модели
res = mod.fit()
# параметры подогнанной модели регрессии
round(res.params, 2)
```
