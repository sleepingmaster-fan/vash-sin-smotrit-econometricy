```python
# подключение библиотек
import numpy as np
import pandas as pd

from statsmodels.formula.api import ols

# Не показывать Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('wage2.csv')
```

```python
# спецификация модели через формулу
mod = ols(formula='np.log(wage)~1+age+IQ+south', data=df)
# подгонка модели
res = mod.fit()
```

```python
res.params
```

```python
# R^2 это доля объяснённой вариации зависимой переменной.
res.rsquared.round(3)
# Модель объясняет 14.6% различий в логарифме заработной платы с помощью переменных age, IQ и south.
```

```python
# adj R^2 он учитывает количество переменных в модели
# учитывает количество объясняющих переменных и показывает, что после корректировки модель объясняет 14.3% вариации зависимой переменной.
res.rsquared_adj
```

```python
# спецификация модели через формулу
mod = ols(formula='np.log(wage)~1+age+IQ+south+urban', data=df)
# подгонка модели
res = mod.fit()
```

```python
# параметры подогнанной модели регрессии с округлением
res.params.round(2)
```

```python
# Создадим датафрейм с новыми данными регрессоров для прогноза
new_data = pd.DataFrame({'age': [39, 27, 23, 32], 'IQ': [115, 106, 113, 102], 'south': [0,1,1,0], 'urban': [1,1, 0, 1] })
new_data
```

```python
# Прогноз
np.exp(res.predict(exog=new_data, transform=True)).round(2)
```
